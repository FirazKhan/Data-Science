### 12. Finding Best budget Restaurants in any given location. Create a function return_budget_resturant, the arguments to this function will be location and rest_type.The function should return a budget resturants in the given loaction. The price of a good budget resturant should be less than 400 and rating should be more than 4.

```python
def return_budget_restaurant(location, rest_type):
    # Assuming you have a DataFrame named 'df' with restaurant data
    # Filter based on location, restaurant type, price, and rating
    filtered_data = df[(df['listed_in(city)'] == location) &
                       (df['listed_in(type)'] == rest_type) &
                       (df['approx_cost(for two people)'] < 400) &
                       (df['rate'] > 4)]

    # Extract restaurant names from the filtered data
    budget_restaurants = filtered_data['name'].tolist()

    return budget_restaurants

# Example usage
result = return_budget_restaurant('HSR', 'Quick Bites')
print(result)
```

```python
def return_budget_restaurant(location, rest_type):
    # Filter the dataset based on location and restaurant type
    filtered_df = df[(df['location'] == location) & (df['rest_type'] == rest_type)]
    
    # Further filter based on cost and rating criteria
    budget_restaurants = filtered_df[(filtered_df['approx_cost(for two people)'] < 400) & (filtered_df['rate'] > 4)]
    
    # Return the names of the budget restaurants
    return budget_restaurants['name'].tolist()
result = return_budget_restaurant('HSR', 'Quick Bites')
print(result)
```

### 13.Which are the most common cuisines in each locations.

```python
# First, we need to clean and preprocess the 'cuisines' and 'location' columns

# Drop any rows where location or cuisines are missing
data_cleaned = df.dropna(subset=['location', 'cuisines'])

# Split the cuisines column into a list of cuisines for each restaurant
data_cleaned['cuisines'] = data_cleaned['cuisines'].str.split(', ')

# Explode the dataframe to have one row per cuisine per restaurant
data_exploded = data_cleaned.explode('cuisines')

# Group by location and cuisine, then count the occurrences
cuisine_counts = data_exploded.groupby(['location', 'cuisines']).size().reset_index(name='count')

# Find the most common cuisine for each location
most_common_cuisines = cuisine_counts.loc[cuisine_counts.groupby('location')['count'].idxmax()]

most_common_cuisines
```
